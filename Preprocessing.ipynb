```python
num_cols=data[['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol']]
num_cols.isna().sum()
```

```python
#Simple imputer
from sklearn.impute import SimpleImputer
#creating instance of package simple imputer od sk learn library
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
#fit numerical data into imputer
imputer=imputer.fit(num_cols)
```

```python
num_cols=imputer.transform(num_cols)
type(num_cols)

```

```python
num_cols=pd.DataFrame(num_cols,columns=[['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol']])
```

```python
num_cols
```

```python
num_cols.isna().sum()
```

```python
data.drop(['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol'],inplace=True,axis=1)
data.head()
```

```python
data=pd.concat([num_cols,data],axis=1)
data.head()
```

```python
#method 2
data2=pd.read_csv('beer-servings.csv',index_col='Unnamed: 0')
data2.isna().sum()
```

```python
#use the float type columns only for plotting
freqGraph=data2.select_dtypes(include=['float'])
freqGraph.hist(figsize=(20,15))
plt.show()
```

```python
#All the columns are right skewed
data2.columns
```

```python
data2['beer_servings']=data2['beer_servings'].fillna(data2['beer_servings'].median())
data2
```

```python
data2.columns
```

```python
for col in ['spirit_servings','wine_servings','total_litres_of_pure_alcohol','beer_servings']:
  data2[col]=data2[col].fillna(data2[col].mean())



```

```python
  data2.isna().sum()
```

```python
#Outliers
plt.boxplot(data2['beer_servings'])
plt.show()
```

```python
#Outliers
plt.boxplot(data2['spirit_servings'])
plt.show()
```

```python
Q1=np.percentile(data2['spirit_servings'],25,interpolation='midpoint')
Q2=np.percentile(data2['spirit_servings'],50,interpolation='midpoint')
Q3=np.percentile(data2['spirit_servings'],75,interpolation='midpoint')
print(Q1,Q2,Q3)
```

```python
data2['spirit_servings'].median()
```

```python
IQR=Q3-Q1
IQR
```

```python
low_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print(low_limit)
print(upper_limit)
```

```python
outlier_list=[]
for i in data2['spirit_servings']:
  if (i>upper_limit) or (i<low_limit):
    outlier_list.append(i)
    
```
